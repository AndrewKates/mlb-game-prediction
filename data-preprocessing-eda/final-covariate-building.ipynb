```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import os

PATH_TO_DATA = '../data/'
PATH_TO_OUTPUT_DATA = '../output-data/'
```

### Execution plan
So essentially we can build our covariates, then save this table. We use the table for x_train, x_test, etc. 

```python
BASEBALL_DATA_DF = pd.read_csv(PATH_TO_OUTPUT_DATA + 'COVARIATE_DATA_PITCHER_BATTER_v3.csv')
BASEBALL_DATA_DF['dt'] = BASEBALL_DATA_DF['dt'].apply(lambda x:
                                                        dt.datetime.strptime(str(x), '%Y-%m-%d'))
BASEBALL_DATA_DF.head(10)
```

```python
ELO = pd.read_csv(PATH_TO_OUTPUT_DATA + 'ELO_ratings.csv')
ELO['dt'] = ELO['dt'].apply(lambda x:
                                                        dt.datetime.strptime(str(x), '%Y-%m-%d'))
ELO
```

```python
BASEBALL_DATA_DF['home_ELO'] = 0
BASEBALL_DATA_DF['away_ELO'] = 0

for i in range(len(BASEBALL_DATA_DF)):
    if (i % 4000 == 0):
        print(i)
    BASEBALL_DATA_DF.loc[i, 'home_ELO'] = ELO[
        ELO['dt'] == BASEBALL_DATA_DF.loc[i, 'dt']][BASEBALL_DATA_DF.loc[i, 'home_team']].values[0]
    BASEBALL_DATA_DF.loc[i, 'away_ELO'] = ELO[
        ELO['dt'] == BASEBALL_DATA_DF.loc[i, 'dt']][BASEBALL_DATA_DF.loc[i, 'away_team']].values[0]
```

```python
BASEBALL_DATA_DF['ELO_difference'] = BASEBALL_DATA_DF['home_ELO'] - BASEBALL_DATA_DF['away_ELO']
```

```python
# SAME FUNCTION FROM EARLIER
def expected_elo_outcome(home_elo_prior, away_elo_prior, EXP_CONSTANT = 200):
    '''
        note that home and away are in AWAY @ HOME order in MLB_GAME_DF_FINAL
        dt = date of the game
        takes both teams current ELOs and computes
        check runtime!
    '''
    elo_expectation = (1 + 10**(-(home_elo_prior - away_elo_prior)/EXP_CONSTANT))**(-1) # equation works
    # function not used with compute_ELOs
    return elo_expectation
```

```python
BASEBALL_DATA_DF['E[ELO]'] = expected_elo_outcome(BASEBALL_DATA_DF['home_ELO'], BASEBALL_DATA_DF['away_ELO'])
```

```python
BASEBALL_DATA_DF['home_rest'] = 0
BASEBALL_DATA_DF['away_rest'] = 0
```

```python
# adding days of rest
# rest_days = dict(zip(list(set(BASEBALL_DATA_DF['home_team'])), [0 for _ in range(30)]))
last_game = dict(zip(list(set(BASEBALL_DATA_DF['home_team'])), [-1 for _ in range(30)]))
```

```python
def home_rest(i):
    home_index = BASEBALL_DATA_DF.loc[i]['home_team']

    try:
        days_rest = (BASEBALL_DATA_DF.loc[i, 'dt'] - last_game[home_index]).days
        last_game[home_index] = BASEBALL_DATA_DF.loc[i, 'dt']
        return days_rest
    except:
        last_game[home_index] = BASEBALL_DATA_DF.loc[i, 'dt']
        return 5 # >= 5

def away_rest(i):
    away_index = BASEBALL_DATA_DF.loc[i]['away_team']

    try:
        days_rest = (BASEBALL_DATA_DF.loc[i, 'dt'] - last_game[away_index]).days
        last_game[away_index] = BASEBALL_DATA_DF.loc[i, 'dt']
        return days_rest
    except:
        last_game[away_index] = BASEBALL_DATA_DF.loc[i, 'dt']
        return 5 # >= 5
```

```python
for i in range(len(BASEBALL_DATA_DF)):
    if (i % 4000 == 0):
        print(i)
    BASEBALL_DATA_DF.loc[i, 'home_rest'] = home_rest(i)
    BASEBALL_DATA_DF.loc[i, 'away_rest'] = away_rest(i)
```
