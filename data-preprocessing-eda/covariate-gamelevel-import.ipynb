```python
import pandas as pd
from retrosheet_scraper import get_retrosheet_data
import time
import os
```

```python
MASTER_DF = pd.DataFrame( 
            columns = ['date', 'game_id', 'team', 'player_id', 'AB', 'RS', 'H', 'HR', 'TB', 'BB', 'SF'])
```

```python
start_time = time.time()
for file in os.listdir('../data/event'):
    MASTER_DF = MASTER_DF.append(get_retrosheet_data(file))
    
final_time = time.time() - start_time
print("--- " + str(round((final_time)*10000)/10000) + " seconds ---")
```

```python
MASTER_DF
```

```python
MASTER_DF = MASTER_DF.sort_values(by = 'date', ascending = True).reset_index(drop = True)
```

```python
len(MASTER_DF[MASTER_DF.game_id != '0'].groupby('date'))
```

```python
import datetime as dt

MASTER_DF['date'] = MASTER_DF['date'].apply(lambda x: dt.date(int(x[0:4]), int(x[4:6]), int(x[6:8])))
```

```python
MASTER_DF.rename({'date': 'dt'}, axis = 1, inplace = True)
```

```python
MASTER_DF['Y'] = (MASTER_DF.dt).apply(lambda x: x.year)
MASTER_DF['M'] = (MASTER_DF.dt).apply(lambda x: x.month)
MASTER_DF['D'] = (MASTER_DF.dt).apply(lambda x: x.day)
```

```python
MASTER_DF = MASTER_DF[['dt', 'Y', 'M', 'D', 'game_id', 'team', 'player_id', 
                       'AB', 'RS', 'H', 'HR', 'TB', 'BB', 'SF']]
```
