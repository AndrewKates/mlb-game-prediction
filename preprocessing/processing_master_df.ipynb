```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

PATH_TO_DATA = '../data/'
PATH_TO_OUTPUT_DATA = '../output-data/'
MASTER_DF = pd.read_csv(PATH_TO_OUTPUT_DATA + 'COVARIATE_MASTER.csv')
```

```python
# get only 2000: MASTER_DF[(MASTER_DF.dt).apply(lambda x: x.year) == 2000]
MASTER_DF
```

```python
# drop 50k irrelevant rows
MASTER_AGG_DF = MASTER_DF.drop(columns=['game_id']).groupby(['team', 'date']).sum().reset_index()
```

```python
MASTER_AGG_DF.to_csv(PATH_TO_OUTPUT_DATA + 'COV_AGGREGATE.csv', index = False)
```

```python
MASTER_AGG_DF['AVG'] = MASTER_AGG_DF['H']/MASTER_AGG_DF['AB']
```
